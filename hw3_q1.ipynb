```python
import pandas as pd

data_file_path = '/content/drive/MyDrive/hw3-data/data.csv'
label_file_path = '/content/drive/MyDrive/hw3-data/label.csv'

data_df = pd.read_csv(data_file_path, header=None)
label_df = pd.read_csv(label_file_path, header=None)
```

```python
import numpy as np
```

```python
combined_df = pd.concat([data_df, label_df], axis=1)
combined_df.head()
```

```python
instances = []
for _, row in combined_df.iterrows():
  instances.append([list(row.iloc[:-1]), row.iloc[-1]])
```

```python
from scipy.spatial import distance

def euclidean_distance(instance1, instance2):
  return distance.euclidean(instance1, instance2)

def cosine_distance(instance1, instance2):
  return distance.cosine(instance1, instance2)

def jaccard_distance(instance1, instance2):
  return distance.jaccard(instance1, instance2)
```

```python
def createEmptyListOfLists(numSubLists):
    myList = []
    for i in range(numSubLists):
        myList.append([])
    return myList

def assign(instance, centroids):
    instance_data = instance[0]
    minDistance = jaccard_distance(instance_data, centroids[0]) # needs to be changed accordingly
    minDistanceIndex = 0
    for i in range(1, len(centroids)):
        d = jaccard_distance(instance_data, centroids[i]). # needs to be changed accordingly
        if (d < minDistance):
            minDistance = d
            minDistanceIndex = i
    return minDistanceIndex


def assignAll(instances, centroids):
    clusters = createEmptyListOfLists(len(centroids))
    for instance in instances:
        clusterIndex = assign(instance, centroids)
        clusters[clusterIndex].append(instance)
    return clusters


def computeCentroids(clusters):
    centroids = []
    for i in range(len(clusters)):
        centroid = meanInstance(clusters[i])
        if centroid is not None:
          centroids.append(centroid)
    return centroids

def meanInstance(instanceList):
    numInstances = len(instanceList)
    if (numInstances == 0):
        return
    numAttributes = len(instanceList[0][0])
    means = [0] * (numAttributes)
    for instance in instanceList:
        for i in range(0, numAttributes):
            means[i] += instance[0][i]
    for i in range(0, numAttributes):
        means[i] /= float(numInstances)
    return tuple(means)

def computeWithinss(clusters, centroids):
    result = 0
    for i in range(len(centroids)):
        centroid = centroids[i]
        cluster = clusters[i]
        for instance in cluster:
            result += jaccard_distance(centroid, instance[0]) # needs to be changed accordingly
    return result
```

```python
## K means
import time, random


def kmeans(instances, k, initCentroids=None):
    result = {}
    if (initCentroids == None or len(initCentroids) < k):
        # randomly select k initial centroids
        random.seed(time.time())
        centroids = [list(x[0]) for x in random.sample(instances, k)]
    else:
        centroids = initCentroids
    prevCentroids = []


    iteration = 1
    while (centroids != prevCentroids):
        clusters = assignAll(instances, centroids)
        prevCentroids = centroids
        withinss = computeWithinss(clusters, centroids)
        print ("iteration: " + str(iteration))
        centroids = computeCentroids(clusters)
        iteration += 1

    result["clusters"] = clusters
    result["centroids"] = centroids
    result["withinss"] = withinss
    return result
```

```python
from collections import Counter
def compute_accuracy(clusters):
  sum_accuracy = 0

  for cluster in clusters:
    data_label = [data_pnt[1] for data_pnt in cluster]
    majority_label, count = Counter(data_label).most_common(1)[0]
    sum_accuracy += count/len(data_label)


  return sum_accuracy/len(clusters)


```

```python
## euclidean
euclidean_res = kmeans(instances, 10)
euclidean_res_sse = euclidean_res['withinss']
euclidean_res_accuracy = compute_accuracy(euclidean_res['clusters'])
```

```python
## Cosine Similarity

cosine_res = kmeans(instances, 10)
cosine_res_sse = cosine_res['withinss']
cosine_res_accuracy = compute_accuracy(cosine_res['clusters'])

print("Cosine res sse" , cosine_res_sse)
print("Cosine res accuracy" , cosine_res_accuracy)
```

```python
## jacard similarity


jacard_res = kmeans(instances, 10)
jacard_res_sse = jacard_res['withinss']
jacard_res_accuracy = compute_accuracy(jacard_res['clusters'])

print("Jacard res sse" , jacard_res_sse)
print("Jacard res accuracy" , jacard_res_accuracy)
```

```python
## increase in sse method

import time, random, math


def kmeans_increase_sse(instances, k, initCentroids=None):
    result = {}
    if (initCentroids == None or len(initCentroids) < k):
        # randomly select k initial centroids
        random.seed(time.time())
        centroids = [list(x[0]) for x in random.sample(instances, k)]
    else:
        centroids = initCentroids

    prevSSE = math.inf

    iteration = 1
    while True:
        clusters = assignAll(instances, centroids)
        currSSE = computeWithinss(clusters, centroids)
        if (currSSE >= prevSSE):
          break
        prevSSE = currSSE
        print ("iteration: " + str(iteration))
        centroids = computeCentroids(clusters)
        iteration += 1

    result["clusters"] = clusters
    result["centroids"] = centroids
    result["withinss"] = currSSE
    return result
```

```python
## euclidean increase in sse
euclidean_res_increase_sse = kmeans_increase_sse(instances, 10)
euclidean_res_sse2 = euclidean_res_increase_sse['withinss']
print("Euclidean res sse" , euclidean_res_sse2)
```

```python
## cosine increase in sse
cosine_res_increase_sse = kmeans_increase_sse(instances, 10)
cosine_res_sse2 = cosine_res_increase_sse['withinss']
print("Cosine res sse" , cosine_res_sse2)
```

```python
## jaccard increase in sse
jaccard_res_increase_sse = kmeans_increase_sse(instances, 10)
jaccard_res_sse2 = jaccard_res_increase_sse['withinss']
print("Jaccard res sse" , jaccard_res_sse2)
```

```python
## K means
import time, random


def kmeans_max_iteraions(instances, k, initCentroids=None):
    result = {}
    max_iterations = 100
    if (initCentroids == None or len(initCentroids) < k):
        # randomly select k initial centroids
        random.seed(time.time())
        centroids = [list(x[0]) for x in random.sample(instances, k)]
    else:
        centroids = initCentroids

    iteration = 1
    while (iteration <= max_iterations):
        clusters = assignAll(instances, centroids)
        withinss = computeWithinss(clusters, centroids)
        print ("iteration: " + str(iteration))
        centroids = computeCentroids(clusters)
        iteration += 1

    result["clusters"] = clusters
    result["centroids"] = centroids
    result["withinss"] = withinss
    return result
```
